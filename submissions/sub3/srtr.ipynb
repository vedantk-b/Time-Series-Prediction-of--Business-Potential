```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("fndip.csv")
```

```python
df.head()
```

```python
col = []
for idx, i in enumerate(df.columns):
    if(idx>0):
        nc = i[6:]
        ncn = int(nc)
        col.append(ncn)
col
dfr = df.drop(columns={"Month"})
col
dfr.loc[len(dfr.index)] = col
dfr
```

```python
dfr = dfr.iloc[:, np.argsort(dfr.iloc[15])]
```

```python
fin = dfr.drop(15)
```

```python
fin["Month"] = ["Month "+str(i) for i in range(73, 88)]
fin.set_index("Month", inplace=True)
```

```python
fin.to_csv("fndipdip.csv")
fin.shape
```
